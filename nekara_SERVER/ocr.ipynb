```python
import easyocr

reader = easyocr.Reader(['ko', 'en'], gpu=False)
```

```python
result = reader.readtext("test.jpg")
img = cv2.imread("test.jpg")

for i in range(len(result)):
  print(result[i])
```

```python
for i in range(len(result)):
  x,y = result[i][0][0]
  w = int(result[i][0][1][0] - x)
  h = int(result[i][0][2][1] - y)
  img = cv2.rectangle(img,(int(x),int(y)),(int(x)+int(w),int(y)+int(h)),(0,0,255),2)
```

```python

cv2.imwrite('test2.jpg', img)
```
